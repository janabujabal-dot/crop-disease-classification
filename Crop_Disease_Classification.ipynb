```python
import torch
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import torchvision.transforms as transforms
import os

IMG_SIZE = 224
BATCH_SIZE = 16

train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
])

valid_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
])

class CropRGBDataset(Dataset):
    def __init__(self, df, image_dir, transform=None):
        self.df = df.reset_index(drop=True)
        self.image_dir = image_dir
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        filename = self.df.loc[idx, "filename"]
        label = self.df.loc[idx, "label_id"]

        image_path = os.path.join(self.image_dir, filename)
        image = Image.open(image_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label, dtype=torch.long)

train_dataset = CropRGBDataset(train_split_df, train_rgb_path, train_transform)
valid_dataset = CropRGBDataset(valid_split_df, train_rgb_path, valid_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=BATCH_SIZE, shuffle=False)

images, labels = next(iter(train_loader))

print(images.shape)
print(labels.shape)
print(labels[:10])
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class FocalLoss(nn.Module):
    def __init__(self, gamma=2.0, alpha=None):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss

        if self.alpha is not None:
            alpha_t = self.alpha[targets]
            loss = alpha_t * loss

        return loss.mean()
```

```python
class CropHSDataset(Dataset):
    def __init__(self, df, image_dir, target_bands=125):
        self.df = df.reset_index(drop=True)
        self.image_dir = image_dir
        self.target_bands = target_bands

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        filename = self.df.loc[idx, "filename"].replace(".png", ".tif")
        label = self.df.loc[idx, "label_id"]

        image_path = os.path.join(self.image_dir, filename)
        image = tiff.imread(image_path).astype(np.float32)

        # توحيد عدد الـ bands
        if image.shape[2] > self.target_bands:
            image = image[:, :, :self.target_bands]
        elif image.shape[2] < self.target_bands:
            pad = self.target_bands - image.shape[2]
            image = np.pad(image, ((0, 0), (0, 0), (0, pad)), mode="constant")

        image = image / 65535.0
        image = torch.tensor(image).permute(2, 0, 1)

        return image, torch.tensor(label, dtype=torch.long)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

class HSResNet18(nn.Module):
    def __init__(self, num_classes=3):
        super().__init__()
        
        self.band_reduction = nn.Conv2d(
            in_channels=125,
            out_channels=3,
            kernel_size=1
        )
        
        self.backbone = models.resnet18(weights=None)
        self.backbone.fc = nn.Linear(self.backbone.fc.in_features, num_classes)

    def forward(self, x):
        x = self.band_reduction(x)
        x = self.backbone(x)
        return x

model = HSResNet18(num_classes=3).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

print(model.band_reduction)
print(model.backbone.fc)
```

```python
class CropHSMSDataset(Dataset):
    def __init__(self, df, hs_dir, ms_dir, target_bands=125):
        self.df = df.reset_index(drop=True)
        self.hs_dir = hs_dir
        self.ms_dir = ms_dir
        self.target_bands = target_bands

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        filename = self.df.loc[idx, "filename"].replace(".png", ".tif")
        label = self.df.loc[idx, "label_id"]

        # HS image
        hs_path = os.path.join(self.hs_dir, filename)
        hs = tiff.imread(hs_path).astype(np.float32)

        if hs.shape[2] > self.target_bands:
            hs = hs[:, :, :self.target_bands]
        elif hs.shape[2] < self.target_bands:
            pad = self.target_bands - hs.shape[2]
            hs = np.pad(hs, ((0, 0), (0, 0), (0, pad)), mode="constant")

        hs = hs / 65535.0
        hs = torch.tensor(hs).permute(2, 0, 1)

        # MS image
        ms_path = os.path.join(self.ms_dir, filename)
        ms = tiff.imread(ms_path).astype(np.float32)
        ms = ms / 65535.0
        ms = torch.tensor(ms).permute(2, 0, 1)

        return hs, ms, torch.tensor(label, dtype=torch.long)
```

```python
import torch
import torch.nn as nn
from torchvision import models

class DualBranchFusion(nn.Module):
    def __init__(self, num_classes=3):
        super().__init__()

        # HS branch: 125 → 3
        self.hs_reduction = nn.Conv2d(125, 3, kernel_size=1)

        self.hs_backbone = models.resnet18(weights=None)
        self.hs_backbone.fc = nn.Identity()

        # MS branch: 5 channels
        self.ms_backbone = models.resnet18(weights=None)
        self.ms_backbone.conv1 = nn.Conv2d(
            5, 64, kernel_size=7, stride=2, padding=3, bias=False
        )
        self.ms_backbone.fc = nn.Identity()

        # Fusion classifier
        self.classifier = nn.Sequential(
            nn.Linear(512 + 512, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, hs, ms):
        hs = self.hs_reduction(hs)
        hs_features = self.hs_backbone(hs)

        ms_features = self.ms_backbone(ms)

        fused = torch.cat([hs_features, ms_features], dim=1)
        out = self.classifier(fused)

        return out

model = DualBranchFusion(num_classes=3).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

print(model)
```

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import models
import tifffile as tiff
from sklearn.model_selection import train_test_split

base_path = "/kaggle/input/competitions/beyond-visible-spectrum-ai-for-agriculture-2026/Kaggle_Prepared"

train_hs_path = os.path.join(base_path, "train", "HS")
train_ms_path = os.path.join(base_path, "train", "MS")
train_rgb_path = os.path.join(base_path, "train", "RGB")

print(os.listdir(os.path.join(base_path, "train")))
```

```python
def get_label(filename):
    return filename.split("_")[0]

train_files = os.listdir(train_rgb_path)

train_df = pd.DataFrame({
    "filename": train_files,
    "label": [get_label(f) for f in train_files]
})

label_map = {
    "Health": 0,
    "Rust": 1,
    "Other": 2
}

train_df["label_id"] = train_df["label"].map(label_map)

train_split_df, valid_split_df = train_test_split(
    train_df,
    test_size=0.2,
    stratify=train_df["label_id"],
    random_state=42
)

print(train_split_df["label"].value_counts())
print(valid_split_df["label"].value_counts())
```

```python
class CropHSDataset(Dataset):
    def __init__(self, df, image_dir, target_bands=125):
        self.df = df.reset_index(drop=True)
        self.image_dir = image_dir
        self.target_bands = target_bands

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        filename = self.df.loc[idx, "filename"].replace(".png", ".tif")
        label = self.df.loc[idx, "label_id"]

        image = tiff.imread(os.path.join(self.image_dir, filename)).astype(np.float32)

        if image.shape[2] > self.target_bands:
            image = image[:, :, :self.target_bands]
        elif image.shape[2] < self.target_bands:
            pad = self.target_bands - image.shape[2]
            image = np.pad(image, ((0, 0), (0, 0), (0, pad)), mode="constant")

        image = image / 65535.0
        image = torch.tensor(image).permute(2, 0, 1)

        return image, torch.tensor(label, dtype=torch.long)

train_hs_loader = DataLoader(CropHSDataset(train_split_df, train_hs_path), batch_size=16, shuffle=True)
valid_hs_loader = DataLoader(CropHSDataset(valid_split_df, train_hs_path), batch_size=16, shuffle=False)

images, labels = next(iter(train_hs_loader))
print(images.shape, labels.shape)
```
